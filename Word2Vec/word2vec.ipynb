# Read data

```python
with open("./data/text8", "r") as fid:
    corpus = fid.readline().strip()
```

```python
corpus[:1000]
```

```python
len(corpus)
```

```python
tokens = corpus.split(" ")
```

```python
len(tokens)
```

Создаём словарь токенов

```python
tokens_dict = {t: i+1 for i, t in enumerate(set(tokens))}
```

```python
tokens_dict
```

```python
tokens_dict.update({" ": 0})
```

```python
idx_dict  = {v: k for k, v in tokens_dict.items()}
```

```python
idx_dict
```

```python
tokens_dict
```

Переводим весь корпус в индексы слов

```python
tokens_idx = [tokens_dict[r] for r in tokens]
```

```python
tokens_idx
```

```python
from pickle import dump
```

```python
with open("corpus.pkl", "wb") as fid:
    dump(tokens_idx, fid)
```
